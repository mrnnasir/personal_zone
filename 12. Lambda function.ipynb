```python
df = sns.load_dataset('diamonds')
df
```

```python
def my_price_perception(price, cut):
    if cut in ['Premium', 'Ideal'] and price < 2000:
        return 'Worth it, buy now!'
    elif price < 1000:
        return 'Buy'
    elif price < 2500:
        return 'Think twice'
    else:
        return 'No thanks' 
```

We create a column '`MyValuePerception`' by using the apply function with a lambda against the row using `axis=1`
  * the argument x for the lambda function is df, our DataFrame
  * for the `my_price_perception` function, we parse the price and the cut.

```python
df['MyValuePerception'] = df.apply(lambda x: my_price_perception(x['price'], x['cut']), axis=1)
df
```

```python
df['MyValuePerception'].value_counts()
```

```python
df_practice = sns.load_dataset('tips')
print(f"DataFrame shape: {df_practice.shape}")
df_practice
```

```python
def day_define(day):
    if day in ['Sat','Sun']:
        return True
    else:
        return False

df_practice['day'] = df_practice.apply(lambda x: day_define(x['day']), axis = 1)
df_practice
```
