```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # used for loading datasets
```

## Bar Plot

You have already seen a variation of the bar plot in the histogram. In a standard bar plot, the x-axis has categories of data, and the bar height is proportional to the values in the dataset. You can then visually compare the values for those categories. We are using the insurance dataset. It has records on the relationship between personal attributes (age, gender, BMI: body mass index, family size, smoking habits), geographic factors, and their impact on medical insurance charges.

```python
df = pd.read_csv('insurance.csv')
df = df.sample(n=50, random_state=1)
df.head(3)
```

We use `plt.bar()` for bar plot. The documentation is found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)
* We are first counting the frequencies with `.value_counts()`, then use` plt.bar()`


```python
categorical_count = df['children'].value_counts()
categorical_count
```

We parse that to `plt.bar()`, where in the x-axis we place the categories name (located in the index) and the height we parse the count

```python
plt.bar(x=categorical_count.index, height=categorical_count)
plt.show()
```

We use `plt.barh()` for bar plot in horizontal mode. The documentation is found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html)
* The arguments are y for the y-axis and width for the categories levels

```python
categorical_count = df['children'].value_counts()
plt.barh(y=categorical_count.index, width=categorical_count)
plt.show()
```

## Pie Chart

A pie plot shows one series of data. The area of the circle denotes 100% of the data. Each slice or wedge of the pie shows the relative percentage for that category. It provides a quick visual summary of the relative size of a category.  We are using the insurance dataset.

```python
df = pd.read_csv('insurance.csv')
df = df.sample(n=50, random_state=1)
df.head(3)
```

We are using `plt.pie()`. The documentation is found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie)
* We are first counting the frequencies with .value_counts(), then use `plt.pie()`. The arguments are x and labels

```python
categorical_count = df.value_counts('region',normalize=True)
plt.pie(x=categorical_count, labels=categorical_count.index)
plt.show()
```

Here, we can easily see that the NorthEast region is the least common among analysed records. 

```python
df_practice = sns.load_dataset('tips')
df_practice = df_practice.sample(n=50, random_state=1)
df_practice.head(10)
```

```python
tips_by_day= df_practice.value_counts('day', normalize=True)
plt.pie(x=tips_by_day, labels=tips_by_day.index )
plt.show()
```

## 3D Plot

For a 3D plot in Matplotlib, we use `plt.axes` to create a plot with three axes. The documentation is found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axes.html)

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
```

We can draw a 3D Line Plot by adding to an axes the method `.plot3D()`, and parsing the 3D arrays to match the x, y and z axis
* First, we create a Figure with plt.Figure. Then a 3D axes with `plt.axes()`
* x, y and z are generated using NumPy mathematical functions. Don't worry about what the functions mean. You may remember sin and cos from school. 
* We plot x, y, and z with `.plot3D()`

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
z = np.linspace(0, 5, 500)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
ax.plot3D(x, y, z)
plt.show()
```

In addition, we can draw a 3D Scatter Plot. The documentation is found [here](https://matplotlib.org/stable/gallery/mplot3d/scatter3d.html)

**Use Case:** These 3D scatter plots are used to display data points in three dimensions, showing the relationship between three continuous variables. This can help identify clusters, outliers, trends, or patterns in multi-dimensional space.

* **When to Use:** Ideal for data exploration in machine learning, particularly in clustering and outlier detection, or any application where the interactions between three variables are of interest.
* **Example Application:** Visualising groups of customers based on three different spending categories in market segmentation.

* We will consider the insurance dataset.

```python
df = pd.read_csv('insurance.csv')
df = df.sample(n=50, random_state=1)
df.head(3)
```

The approach is similar to the previous examples, but here we plot by using `.scatter() `and parse the x, y and z data.

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')

x = df['age']
y = df['bmi']
z = df['expenses']

ax.scatter(x, y, z)
plt.show()
```

## Images

Matplotlib can handle image datasets, as you saw in the NumPy lesson. You can read the data as a NumPy array using plt.imread(). The documentation is found here
First, you will need your image location. In this case, it is a URL. You can ignore the matplotlib warning: "Passing a URL is deprecated" and suggest using another library for managing the task of loading URL images. In the course, we will use this command to load images from the local directory.
the second argument is optional and is the format. If not given, the format is deduced from the filename. If nothing can be deduced, PNG is assumed.

```python
image_url = "http://ci-jshint.herokuapp.com/static/images/logo.png"
image = plt.imread(image_url, format='png')
```

```python
plt.imshow(image)
plt.show()
```

You can plot the image using `plt.imshow()`. The documentation is found [here](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html)
