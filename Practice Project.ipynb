```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer_shopping_data.csv', index_col=None)
print(df.shape)
df.head()
```

```python
df.columns
```

```python
# Fixed spaces and column names

df.rename(columns= {' age ': 'age',
           'invoice_date': 'Datetime'}, inplace=True)
df
```

```python
# Converted Datetime column to Datetime format
# May need to try: df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
df['Datetime'] = pd.to_datetime(df['Datetime'], errors='raise', format='mixed')
df
```

```python
df[df['Datetime'].isnull()]
```

```python
df.info()
```

```python
# Dropped all NaN values

df= df.dropna()
df
```

```python
df.isnull().sum()
```

```python
# Randomly selected 5000 data
# Created separate.csv file and set Index = False

np.random.seed(1)
df = df.sample(n=5000, random_state=1)
print(df.shape)
df.head()
df.to_csv('customer_shopping_data(resampled).csv', index=False)
```
