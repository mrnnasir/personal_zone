```python
# Removed duplicate 'Transaction_ID','User_ID' data

df = df.drop_duplicates(subset=['Transaction_ID','User_ID'], keep='first')
df
```

```python
df = df.drop(['Account_Age', 'Number_of_Transactions_Last_24H'], axis=1)
df
```
